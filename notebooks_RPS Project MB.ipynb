```python
def RPS_GAME():
    print('--- Play Rock Paper Scissors! ---')

    valid_moves=['rock', 'paper', 'scissors']

    playing = True
    while playing:
        player1=input("Player 1 - Enter rock, paper, or scissors: ").lower()
        player2=input("Player 2 - Enter rock, paper, or scissors: ").lower()
    
        moves={
            'tie': 'It is a tie! No winner.',
            'rock vs paper': 'Paper beats rock. Player 2 wins!',
            'paper vs rock': 'Paper beats rock. Player 1 wins!',
            'rock vs scissors': 'Rock beats scissors. Player 1 wins!',
            'scissors vs rock': 'Rock beats scissors. Player 2 wins!',
            'paper vs scissors': 'Scissors beat paper. Player 2 wins!',
            'scissors vs paper': 'Scissors beat paper. Player 1 wins!',
            } 

        if player1 not in  valid_moves or player2 not in valid_moves:
            print("Error, please enter a valid move: rock, paper, or scissors.")
        
        if player1 == player2:
            print(f"{moves['tie']}")
        elif player1 == 'rock' and player2 == 'paper':
            print(f"{moves['rock vs paper']}")
        elif player1 == 'paper' and player2 == 'rock':
            print(f"{moves['paper vs rock']}")
        elif player1 == 'rock' and player2 == 'scissors':
            print(f"{moves['rock vs scissors']}")
        elif player1 == 'scissors' and player2 == 'rock':
            print(f"{moves['scissors vs rock']}")
        elif player1 == 'paper' and player2 == 'scissors':
            print(f"{moves['paper vs scissors']}")
        elif player1 == 'scissors' and player2 == 'paper':
            print(f"{moves['scissors vs paper']}")       
        else:
            print("Error, try again.")

        again = input('Play again? (y - yes / n - no)').lower()
        if again !='y':
            print('Thank you for playing!')
            playing = False

RPS_GAME()
```
